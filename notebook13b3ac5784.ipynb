```python
import time
import csv

name = input("Enter your name: ")

filename = name + "_entries.csv"
file = open(filename, 'w', newline='')
writer = csv.writer(file)
writer.writerow(["Name", "Timestamp", "Time Delay", "Number Entered"])

last_time = time.time()

while True:
    entry = input("Enter a number between 0-9 (or type 'quit' to stop): ")
    
    if entry.lower() == 'quit':
        break

    if not entry.isdigit() or not (0 <= int(entry) <= 9):
        print("Please enter a valid number between 0 and 9.")
        continue

    now = time.time()
    delay = now - last_time
    last_time = now

    timestamp = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(now))

    writer.writerow([name, timestamp, round(delay, 2), entry])

file.close()
print("All data saved. Goodbye!")
```
